```python
import pandas as pd
import sklearn
```

```python
print(sklearn.__version__)
```

```python
df = pd.read_csv('./data/cardio_processed_data.csv')
```

```python
df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

```python
# seperate target column and featuers
x = df.drop("cardio" , axis=1)
x.head()
```

```python
y = df['cardio']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.scatter(df['ap_hi'] , df['cardio'])
# plt.scatter(df['height'] , df['cardio'])
# plt.scatter(df['bmi'] , df['cardio'])
plt.scatter(df['cholesterol'] , df['cardio'])
```

# Logisitc Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
log_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])

log_pipeline
```

```python
log_pipeline.fit(x_train,y_train)
```

```python
y_test.head()
```

```python
y_pred = log_pipeline.predict(x_test)
y_pred
```

```python
y_prob = log_pipeline.predict_proba(x_test)[:,1]
y_prob
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
print(accuracy_score(y_test , y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prob)
auc_log = roc_auc_score(y_test, y_prob)

plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {auc_log:.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve – Logistic Regression")
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_pipeline = Pipeline([
    ("scaler", StandardScaler()),   # optional but safe
    ("model", RandomForestClassifier(
        n_estimators=200,
        max_depth=10,
        random_state=42
    ))
])
```

```python
rf_pipeline.fit(x_train, y_train)
```

```python
y_pred_2 = rf_pipeline.predict(x_test)
y_prob_2 = rf_pipeline.predict_proba(x_test)[:,1]

print(accuracy_score(y_test,y_pred_2))
```

```python
print(classification_report(y_test, y_pred_2))
```

```python
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_2)
auc_rf = roc_auc_score(y_test, y_prob_2)

plt.figure()
plt.plot(fpr_rf, tpr_rf, label=f"AUC = {auc_rf:.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve – Random Forest")
plt.legend()
plt.show()
```

```python
print("Logistic Regression AUC:", auc_log)
print("Random Forest AUC:", auc_rf)
```
